```python
#실습 1번
import csv
f=open('temperature.csv', 'r' ,encoding='cp949')
data=csv.reader(f)
next(data)

list_temp=[]
for date, branch, average, low, high in data:
    low=float(low)
    high=float(high)
    list_temp.append( (date, low, high) )
list_temp.sort(key = lambda x: x[1])
low_date=list_temp[0][0]
list_temp.sort(key = lambda x: x[2])
high_date=list_temp[-1][0]

list_temp.sort(key= lambda x: x[2]-x[1])
diff_date=list_temp[-1][0]

print("일교차가 가장 큰 날\n"+ diff_date)
print("최저기온이 가장 낮은 날\n"+ low_date)
print("최고기온이 가장 높은 날\n"+ high_date)

```

```python
#실습 2번

import csv
f=open('subway.csv', 'r', encoding='UTF-8')
data=csv.reader(f)
next(data)


list_name_dict={}
#노선 data[1] 별로 data[4] data[5] 더해서 가장 크고 작은거 3개씩
for month, line_name, station_id, station_name,passenger_in, passenger_out  in data:
    if line_name not in list_name_dict: #키가 없으면
        list_name_dict[line_name]=[] #키값을 리스트로 생성
    passenger_in=int(passenger_in.replace(',' , ''))
    passenger_out=int(passenger_out.replace(',' , ''))
    
    list_name_dict[line_name].append((station_name, passenger_in+passenger_out))
    
for line_name in list_name_dict:
    list_name_dict[line_name].sort(key=lambda x:x[1])
    print(line_name)
    print("이용객이 가장 많은 역 순위")
    for i in range(3):
        name= list_name_dict[line_name][-i-1][0]
        passenger= list_name_dict[line_name][-i-1][1]
        print(i+1,'.',name, passenger)
    print("\n이용객이 가장 적은 역 순위")
    for i in range(3):
        name= list_name_dict[line_name][i][0]
        passenger= list_name_dict[line_name][i][1]
        print(i+1,'.',name, passenger)
    print('\n')

    
```
